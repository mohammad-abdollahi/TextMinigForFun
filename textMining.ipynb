```python
import numpy as np
import pandas as pd 
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
text = pd.read_csv("text_mining.csv").drop(["document_id"],axis=1)
text = text[text['direction']==1]
tex2 = text[(text['category_id']==1) | (text['category_id']==8)]
```

```python
text.head()
```

```python
cliCon = text[text['category_id']==1]
politCon = text[text['category_id']==8]
```

```python
cliCon.head()
```

```python
cli = ''
for i in cliCon.text.as_matrix(): cli = cli + ' ' + i
pol = ''
for i in politCon.text.as_matrix(): pol = pol + ' ' + i
```

```python
from hazm import stopwords_list,Stemmer
stm = Stemmer()
sw = stopwords_list()
swFilt = []
for i in sw:
    i = i.encode("utf-8")
    swFilt.append(i)
print swFilt
```

```python
filt = ""
w = cli.split()
for i in w:
    if i not in swFilt and i != "است." and i != '</br>':
        filt = filt + i +" "
w = filt.split()
cli_filt = ""
for i in w:
    i = i.decode("utf-8")
    j = stm.stem(i)
    j.encode("utf-8")
    cli_filt = cli_filt + j + " "
print cli_filt 
```

```python
filt = ""
w = pol.split()
for i in w:
    if i not in swFilt and i != "است." and i != '</br>':
        filt = filt + i +" "
w = filt.split()
pol_filt = ""
for i in w:
    i = i.decode("utf-8")
    j = stm.stem(i)
    j.encode("utf-8")
    pol_filt = pol_filt + j + " "
print pol_filt 
```

```python
content = [cli_filt, pol_filt]
```

```python
cv = CountVectorizer()
cv.fit(content)
cv_doc = cv.transform(content)
```

```python
df = pd.DataFrame(cv_doc.toarray().transpose(), index = cv.get_feature_names())
```

```python
df.sort_values(0, ascending=False)
```

```python
tfidf = TfidfVectorizer()
tfidf_doc = tfidf.fit_transform(content)
```

```python
df2 = pd.DataFrame(tfidf_doc.toarray().transpose(), index =tfidf.get_feature_names())
```

```python
df2.sort_values(0, ascending=False)
```

```python
tmp = []
for i in tex2['text']:
    i = i.split()
    filt=""
    for j in i:
        if j not in swFilt:
            filt = filt + j +" "
    tmp.append(filt)
tex2['text'] = tmp 
print tex2.head()
from sklearn.model_selection import train_test_split
y, z = train_test_split(tex2, test_size=.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(tfidf.fit_transform(y.text).toarray(), y.category_id)
```

```python
from sklearn.metrics import classification_report
print(classification_report(z.category_id, model.predict(tfidf.transform(z.text).toarray())))
```

```python
from sklearn.tree import DecisionTreeClassifier
tr = DecisionTreeClassifier()
tr.fit(tfidf.fit_transform(y.text).toarray(), y.category_id)
print(classification_report(z.category_id, tr.predict(tfidf.transform(z.text).toarray())))
```

```python
x = text
tmp = []
for i in x['text']:
    if (isinstance(i, float) == False ):
        i = i.split()
        filt=""
        for j in i:
            if j not in swFilt and j != "است.":
                filt = filt + j +" "
        tmp.append(filt)
    else:
        tmp.append("")
x['text'] = tmp 
print x.head()
y, z = train_test_split(x, test_size=.2)
```
